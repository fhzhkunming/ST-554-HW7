#### Fit multiple linear regression (MLR) models
– At least one should include interaction terms \
– At least one should include some polynomial terms (you may want to  standardize your predictors but that is up to you)\
– Use CV to select your best MLR model

```python
# 1. full model
cv_full_model = cross_validate(
    LinearRegression(),               # create a reg object
    X_train,
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")

# 2. reduce model
cv_mlr1 = cross_validate(
    LinearRegression(),
    X_train[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'sulphates']],
    y_train,
    cv=5,
    scoring="neg_mean_squared_error")

# 3. interaction model
X_train_inter = X_train.copy() # make a copy of training set
# create an interaction term
X_train_inter['sulphates_pH'] = X_train_inter['sulphates'] * X_train_inter['pH']

cv_mlr2 = cross_validate(
    LinearRegression(),
    X_train_inter[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'sulphates', 'sulphates_pH']],
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")

# 4. model with polynomial terms
X_train_ploy = X_train.copy() # make a copy of training set
# make a squared term for pH
X_train_ploy['pH_sq'] = X_train_ploy['pH'] ** 2

cv_mlr3 = cross_validate(
    LinearRegression(),
    X_train_ploy[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'pH_sq']],
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")
```

- Choose the best model


```python
# calculate the RMSE of each model
rmse_full = np.sqrt(-cv_full_model['test_score'].mean())
rmse_mlr1 = np.sqrt(-cv_mlr1['test_score'].mean())
rmse_mlr2 = np.sqrt(-cv_mlr2['test_score'].mean())
rmse_mlr3 = np.sqrt(-cv_mlr3['test_score'].mean())
print(rmse_full, rmse_mlr1, rmse_mlr2, rmse_mlr3) # display the valule of RMSEs
```

Based on the RMSE values, the full model shows the best performance, since it has the lowest error compared to the other models.

- Define a function to get the best mean negative log-loss across folds for the best C chosen during cross validation for that specific model version.

```python
# function to get the best mean negative log-loss across folds
def best_cv_score(log_reg_model, positive_class = 1):
    means = log_reg_model.scores_[positive_class].mean(axis = 0)
    return means.max()
```
