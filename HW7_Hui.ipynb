## Read in and Combine Data
- Read in the winequality-red.csv and winequality-white.csv files available on the [UCI machine learning repository site](https://archive.ics.uci.edu/dataset/186/wine+quality).


```python
# import pandas
import pandas as pd
# read in data
wine_red = pd.read_csv('winequality-red.csv', sep=';')
wine_white = pd.read_csv('winequality-white.csv', sep=';')
# check the info of the dataset
wine_red.info()
```

- Combine these two datasets and create a new variable that represents the type of wine (red or white)\
The two data sets have identical structure but have different number of observations, so I combined them and create a new variable 'type' to keep track of which wine is which.

```python
# combine the two datasets and create a new variable 'type' to keep track of which wine is which
wine_red['type'] = 'red'
wine_white['type'] = 'white'
wine = pd.concat([wine_red, wine_white], axis=0, ignore_index=True)
```

```python
wine.info()  # check dataset info of the combined data
```

Rename some of the variables by replacing spaces with underscores for easier modeling

```python
# rename columns by replacing spaces with underscores
wine = wine.rename(columns = lambda x: x.replace(' ', '_'))
# check the info of the renamed dataset
wine.info()
```

## Split the Data
- Split up the data set into a training and test set. For this, use stratified sampling to make sure that you have a similar proportion of white and red wines in the training and test sets. This can be done with the `train_test_split()` function ([see the help](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)).

```python
# import modules needed
import numpy as np
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, LassoCV, Lasso, RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import ElasticNetCV
```

```python
# split the data using stratified sampling
X_train, X_test, y_train, y_test = train_test_split(
    wine.drop(['alcohol','type'], axis= 1),
    wine['alcohol'],                # use 'alcohol' as response
    test_size = 0.2,                # 80/20 split
    random_state = 42,              # set random state for reproducibility
    stratify = wine['type'])        # stratified sampling based on 'type'
```

- Standardize variables. \
I first compute the mean and standard deviation for each numeric predictor using the training data. Next, I standardize these predictors by subtracting their mean and dividing by their standard deviation so that each has mean 0 and standard deviation 1. I then save these training-set means and standard deviations so I can apply the same transformation to the test set.

```python
# calculate the means of numeric variables
means = X_train.apply(np.mean, axis= 0)
means
```

- Fitting multiple linear regression (MLR) models

```python
# full model
cv_full_model = cross_validate(
    LinearRegression(),                     # create a reg oject
    X_train,
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")

# reduce model
cv_mlr1 = cross_validate(
    LinearRegression(),
    X_train[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'sulphates']],
    y_train,
    cv=5,
    scoring="neg_mean_squared_error")

# model with interaction
X_train_inter = X_train.copy()
X_train_inter['sulphates_pH'] = X_train_inter['sulphates'] * X_train_inter['pH']

cv_mlr2 = cross_validate(
    LinearRegression(),
    X_train_inter[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'sulphates', 'sulphates_pH']],
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")

# model with polynomial terms
X_train_ploy = X_train.copy()
X_train_ploy['pH_sq'] = X_train_ploy['pH'] ** 2

cv_mlr3 = cross_validate(
    LinearRegression(),
    X_train_ploy[['fixed_acidity', 'residual_sugar', 'chlorides', 'pH', 'pH_sq']],
    y_train,
    cv = 5,
    scoring = "neg_mean_squared_error")
```

- Choose the best model


```python
# calculate the RMSE of each model
rmse_full = np.sqrt(-cv_full_model['test_score'].mean())
rmse_mlr1 = np.sqrt(-cv_mlr1['test_score'].mean())
rmse_mlr2 = np.sqrt(-cv_mlr2['test_score'].mean())
rmse_mlr3 = np.sqrt(-cv_mlr3['test_score'].mean())
print(rmse_full, rmse_mlr1, rmse_mlr2, rmse_mlr3) # display the valule of RMSEs
```

It looks like the full modle is the best based on RMSE.

- Fit a LASSO model with a set of predictors of your choosing\
    – Use at least five predictors\
    – Use CV to select the tuning parameter

```python
lasso_wine = LassoCV(cv =5, random_state = 10)\
                    .fit(X_train[['fixed_acidity', 'residual_sugar', 'chlorides', 'total_sulfur_dioxide', 'sulphates', 'pH']],
                    y_train)
print(lasso_wine.alpha_) # extracting the selected tuning parameter alpha_
```

This is the tuning parameter α selected by LassoCV. The number is the result of the cross-validation process LASSO uses to balance model fit and coefficient shrinkage.

```python
# display the fitted LASSO coefficients
pd.Series(lasso_wine.coef_,
          index=['fixed_acidity', 'residual_sugar', 'chlorides',
                 'total_sulfur_dioxide', 'sulphates', 'pH'])
```

The coefficients indicate how each predictor is associated with alcohol content after LASSO shrinkage. In this model, higher values of all predictors except `sulphates` are associated with lower alcohol content when the other variables are held constant. `Chlorides` and `residual_sugar` show the strongest negative associations, suggesting they contribute most to lowering predicted alcohol levels. `Sulphates` has a small positive coefficient, indicating a slight positive association with alcohol content relative to the other predictors.

- Fit a Ridge Regression model with a set of predictors of your choosing\
    – Use at least five predictors\
    – Use CV to select the tuning parameter

```python
# set up tuning-partmeter grid
alphas = np.logspace(-3, 3, 100)

# set up a pipeline with RidgeCV
ridge_cv = make_pipeline(
    StandardScaler(),
    RidgeCV(alphas = alphas, cv = 5))

# fit the model
ridge_cv.fit(X_train[['fixed_acidity', 'residual_sugar', 'chlorides', 'total_sulfur_dioxide', 'sulphates']],
            y_train)

# extract the chosen alpha
print(ridge_cv.named_steps['ridgecv'].alpha_)
```

Cross-validation selected a tuning parameter of α = 61.36 for the Ridge model, indicating that the model performs best with relatively strong L2 regularization. This level of shrinkage helps stabilize the coefficients in the presence of correlated predictors.

```python
# check the Ridge coefficients
ridge_cv.named_steps['ridgecv'].coef_
```

The Ridge coefficients show that all predictors except `sulphates` are negatively associated with alcohol content, holding the others constant. `Residual sugar` and `chlorides` have the largest negative coefficients, indicating the strongest associations with lower alcohol levels. `Sulphates` has a small positive coefficient, suggesting a slight positive relationship with alcohol content.

- Fit an Elastic Net model with a set of predictors of your choosing\
    – Use at least five predictors\
    – Use CV to select the tuning parameters

```python
wine_en = make_pipeline(
    StandardScaler(),
    ElasticNetCV(cv = 5, random_state = 10,
                l1_ratio = [0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.96, 0.98, 0.99, 1],
                n_alphas = 50))
wine_en.fit(X_train[['fixed_acidity', 'residual_sugar', 'chlorides',
                     'total_sulfur_dioxide', 'sulphates', 'pH']],
            y_train)
```

```python
en = wine_en.named_steps['elasticnetcv'] # extract the ElasticNetCV model from the pipeline
print(en.alpha_)                         # optimal alpha selected by cross‑validation
print(en.l1_ratio_)                      # optimal L1/L2 mixing parameter selected by CV
print(en.coef_)                          # final Elastic Net coefficients
```

Cross-validation selected α = 0.0065 and an l1_ratio of 0.1, indicating that the Elastic Net model favors a penalty dominated by L2 regularization. The resulting coefficients show that all predictors except `sulphates` are negatively associated with alcohol content. `Residual sugar` and `chlorides` have the largest coefficient magnitudes, suggesting the strongest relationships with alcohol content, while `sulphates` shows a small positive association.
